```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
file_path = 'D:\ml intern\clg_work\XRD\XRD_csv\CCM-14D 1.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract the 'Angle' and 'Intensity' columns
angle = df['Angle']
intensity = df['Intensity']

# Create the graph
plt.plot(angle, intensity)
plt.xlabel('Angle')
plt.ylabel('Intensity')
plt.title('XRD Data')
plt.show()
```
